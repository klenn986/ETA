# Experiment with SciKit Learn

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus.para_str)
```

```python
V = pd.DataFrame(vectorizer.get_feature_names(), columns=['term_str'])
V.index.name = 'term_id'
```

```python
DTM = pd.DataFrame(X.toarray())
```

```python
DTM.index = corpus.index
```
